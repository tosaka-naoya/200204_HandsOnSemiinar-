## 17:45-18:30 Azure Machine Learningの「デザイナー」を使って機械学習モデルを学習させる

### 目的とゴール

- 目的：ノンコードで機械学習モデルを学習させることで、機械学習を実務に活用するイメージを持っていただく
- ゴール：コードを記述しなくても、ドラッグ＆ドロップによるビジュアルインターフェイスを使用して回帰モデルを学習させる

## はじめに

- "回帰" は、項目の "特徴" に基づいて数値 "ラベル" を予測するために使用される機械学習の1つの手法
- たとえば、自動車販売会社では、自動車の特性 (エンジン サイズ、座席数、走行距離など) を使用して、販売価格を予測することができる
- この場合、自動車の特性が特徴であり、販売価格がラベルとなる

- Microsoft Azure Machine Learning デザイナーを使用すると、コードを記述しなくても、ドラッグ＆ドロップによるビジュアルインターフェイスを使用して回帰モデルを作成できる
- 本章では、Azure Machine Learning デザイナーを使用して回帰モデルをトレーニングする
- 推論に回帰モデルを使用する

### デザイナーを使用するための準備

- 作成したAzureびワークスペースからAzure Machine Learning Studioへ移動する
- Designerへ移動する

## データの探索

- 回帰モデルをトレーニングするには、過去の "特徴" (予測を行う対象のエンティティの特性) と既知の "ラベル" 値 (予測するモデルをトレーニングする数値) を含むデータセットが必要

### パイプラインを作成する

- Azure Machine Learning デザイナーを使用するには、機械学習モデルのトレーニングに使用する "パイプライン" を作成する
- このパイプラインは、モデルのトレーニングに使用するデータセットから始まる
- Azure Machine Learning Studio  で、([作成者] の下の) [デザイナー] ページを表示し、[+] を選択して新しいパイプラインを作成する
- [設定] ペインで、既定のパイプライン名 (Pipeline-Created-on- date**) を Auto Price Training に変更する ([設定] ペインが表示されていない場合は、上部のパイプライン名の横にある ⚙ アイコンを選択する
- パイプラインを実行するコンピューティング先を指定する必要があることを確認する
- [設定] ペインで、[コンピューティング先を選択します] を使用して、前に作成したコンピューティング クラスターを選択する

### トレーニング モジュールを追加する
- データのサブセットを使用してモデルをトレーニングする一方で、トレーニング済みモデルのテストに使用するためにデータの一部を確保しておくことが一般的
- これにより、モデルによって予測されるラベルを元のデータセット内の実際の既知のラベルと比較できる

### 次の手順

- 必要なモジュールを追加して構成するときに、上記の参照用イメージを使用する
- 前のユニットで作成した Auto Price Training パイプラインを開く (まだ開いていない場合)。
- 左側のペインの Data Transformations(データ変換) セクションで、Split Data(データの分割) モジュールをキャンバスの Normalize Data(データの正規化) モジュールの下にドラッグする
- 次に、Normalize Data(データの正規化) モジュールの "変換されたデータセット" (左側) の出力をSplit Data(データの分割) モジュールの入力に接続する
- Split Data(データの分割) モジュールを選択し、その設定を次のように構成する
  - Splitting mode(分割モード): Split Rows (行を分割)
  - Fraction of rows in the first output dataset(最初の出力データセット内の行の割合): 0.7
  - [ランダム シード]: 123
  - Stratified split(層化分割): False
- 左側のペインにある [Model Training](モデル トレーニング) セクションを展開し、[モデルのトレーニング] モジュールをキャンバスの [データの分割] モジュールの下にドラッグする
- 次に、[データの分割] モジュールの "結果データセット 1" (左側) の出力を [モデルのトレーニング] モジュールの "データセット" (右側) の入力に接続する
- トレーニングするモデルでは price の値が予測されるため、[モデルのトレーニング] モジュールを選択し、ラベル列 を price に設定するように設定を変更する (大文字と小文字の区別とスペルは正確に一致させる)。
- モデルによって予測される price ラベルは数値であるため、"回帰" アルゴリズムを使用してモデルをトレーニングする必要がある
- Machine Learning Algorithms(機械学習アルゴリズム) セクションを展開し、[回帰] の下で、[線形回帰] モジュールをキャンバスの [データの分割] モジュールの左、[モデルのトレーニング] モジュールの上にドラッグします
- 次に、その出力を、[モデルのトレーニング] モジュールの Untrained model (左側) の入力に接続する

----
- ＜注意＞回帰モデルのトレーニングに使用できるアルゴリズムは複数あるが、 1 つを選択する方法については、[Azure Machine Learning デザイナーの機械学習アルゴリズム チート シート](https://docs.microsoft.com/en-us/azure/machine-learning/algorithm-cheat-sheet)を参照してください。

## 18:45-19:45 Azure Machine Learningの「デザイナー」を使って機械学習モデルで推論する

### 目的とゴール

- 目的:ノンコードで機械学習モデルで推論することで、機械学習を実務に活用するイメージを持っていただく
- ゴール：コードを記述しなくても、ドラッグ＆ドロップによるビジュアルインターフェイスを使用して回帰モデルで推論する

## 回帰モデルを評価する

- 回帰モデルを評価するには、予測ラベルを、トレーニング中に確保しておいた検証データセット内の実際のラベルと比較できますが、これは不正確なプロセスであり、複数のモデルのパフォーマンスを比較するために使用できる単純なメトリックは提供されない

### Evaluate Mode(モデル評価) モジュールを追加する

- 前のユニットで作成した Auto Price Training パイプラインを開く (まだ開いていない場合)
- 左側のペインの Model Scoring & Evaluation　(モデル スコアリングおよび評価) セクションで、Evaluate Mode（モデル評価) モジュールをキャンバスの Score Model(モデルのスコア付け) モジュールの下にドラッグし、Score Model(モデルのスコア付け) モジュールの出力を、Evaluate Mode(モデル評価) モジュールの スコア付けされたデータセット (左側) に接続する
- パイプラインが次のようになっていることを確認する
スクリーンショット 2020-12-06 16.11.50.png

- [送信] を選択し、auto-price-training という名前の既存の実験を使用してパイプラインを実行する
- 実験の実行が完了するまで待つ
---
- 実験の実行が完了したら、Evaluate Mode(モデル評価) モジュールを選択し、[設定] ペインの [出力 + ログ] タブで、[評価結果] セクションの Data outputs(データ出力) の下の [視覚化] アイコンを使用して結果を表示する
- これらには、次の回帰パフォーマンス メトリックが含まれる
  - 平均絶対誤差 (MAE):予測値と真の値の差の平均（この値は、ラベルと同じ単位 (この例ではドル) に基づいており、この値が小さいほど、モデルの予測が正確）
  - 二乗平均平方根誤差 (RMSE):予測と真の値の平均二乗の平方根の差（結果は、ラベルと同じ単位 (ドル) に基づくメトリックになり、MAE (上記) と比較すると、差が大きいほど個々の誤差の分散が大きくなることを示す (たとえば、一部の誤差が非常に小さく、他の誤差が大きい場合など)）
  - 相対二乗誤差 (RSE):予測と真の値の差の 2 乗に基づく、0 から 1 の間の相対的なメトリック（このメトリックが 0 に近いほど、モデルのパフォーマンスが高くなり、このメトリックは相対的なものであるため、ラベルの単位が異なるモデルを比較するために使用できる
  - 相対絶対誤差 (RAE):予測と真の値の絶対差に基づく、0 から 1 の間の相対的なメトリック（このメトリックが 0 に近いほど、モデルのパフォーマンスが高くなり、RSE と同様に、このメトリックは、ラベルの単位が異なるモデルを比較するために使用できる）
  - 決定係数 (R 2):このメトリックは、通常は "R-2 乗" と呼ばれ、予測と真の値の間の分散の量がモデルによって説明される方法を要約する（この値が 1 に近いほど、モデルのパフォーマンスが高くなる）
- Evaluate Model result visualization(モデルの評価結果の視覚化) ウィンドウを閉じる
---
- 異なる回帰アルゴリズムを試して結果を比較するには、[データの分割] モジュールから 2 番目の [モデルのトレーニング] モジュール (異なるアルゴリズムを使用) と 2 番目の [モデルのスコア付け] モジュールに同じ出力を接続する
- 次に、両方の [モデルのスコア付け] モジュールの出力を同じ Evaluate Model(モデル評価) モジュールに接続し、並べて比較する
- ニーズに合った評価メトリックでモデルを特定したら、そのモデルを新しいデータで使用する準備を行うことができる
